```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#unique values in each column
unique_values = covid19.nunique()
print(len(unique_values))
print(unique_values)
```

```python
# drop unnecessary column from dataset
covid19 = covid19.drop(['Suppress'] ,axis=1)
```

```python
covid19
```

```python
column_data_types = covid19.dtypes
print(column_data_types)
```

```python
covid19['Week Ending Date']=covid19['Week Ending Date'].astype('category')
covid19['State']=covid19['State'].astype('category')
```

```python
covid19.info()
```

```python
covid19.isnull()
```

```python
covid19=covid19.dropna()
covid19
```

```python
# Rename columns
covid19= covid19.rename(columns={'Week Ending Date': 'Week_Ending_Date', 'State': 'state', 'Observed Number': 'Observed_Number','Upper Bound Threshold':'Upper_Bound_Threshold','Exceeds Threshold':'Exceeds_Threshold','Average Expected Count':'Average_Expected_Count','Excess Estimate':'Excess_Estimate','Total Excess Estimate':'Total_Excess_Estimate','Percent Excess Estimate	':'Percent_Excess_Estimate','Year':'year','Type':'type','Outcome':'outcome'})
```

```python
covid19.rename
```

```python
covid19['Average Excess Estimate'] = covid19['Total_Excess_Estimate'] - covid19['Excess_Estimate']
covid19
```

```python
duplicate = covid19.duplicated()
print(duplicate)
```
